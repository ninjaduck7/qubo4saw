```python
import matplotlib.pyplot as plt

# Create a 6x4 grid of points
fig, ax = plt.subplots()
grid_width, grid_height = 6, 4

# Draw the grid points
for x in range(grid_width):
    for y in range(grid_height):
        ax.plot(x, y, 'o', color='black')  # Plot each grid point as a small circle

# Define the walk path with arrows
path = [(2, 3), (3, 3), (4, 3), (4, 2)]
for i in range(len(path) - 1):
    start, end = path[i], path[i + 1]
    ax.annotate('', xy=end, xycoords='data', xytext=start, textcoords='data',
                arrowprops=dict(arrowstyle="->", color='red', lw=2))

# Label specific points with improved positioning
labels = {'A': (4, 2), 'B': (3, 2), 'C': (5, 2), 'D': (4, 1)}
offsets = {'A': (0.2, 0.3), 'B': (0.2, 0.3), 'C': (0.3, -0.3), 'D': (-0.3, -0.3)}

for label, coord in labels.items():
    x_offset, y_offset = offsets[label]
    ax.text(coord[0] + x_offset, coord[1] , label, fontsize=12, ha='center', va='center', color='blue')

# Set limits, aspect ratio, and remove axes
ax.set_xlim(-0.5, grid_width - 0.5)
ax.set_ylim(-0.5, grid_height - 0.5)
ax.set_aspect('equal')
ax.axis('off')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_walk(walk, ax, title, success=True):
    # Plotting the walk
    ax.plot(walk[:, 0], walk[:, 1], 'o-', label='Walk')
    ax.set_title(title)
    if success:
        ax.set_facecolor('lightgreen')
    else:
        ax.set_facecolor('salmon')
    # Setting the grid
    ax.grid(True)
    ax.set_xticks(np.arange(-1, 6))
    ax.set_yticks(np.arange(-1, 6))
    ax.axis([-1, 6, -1, 6])

def main():
    # Initial walk coordinates
    initial_walk = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])

    # Success: Apply a symmetry operation that remains valid
    successful_walk = np.array([[1, 1], [1, 2], [2, 2], [3, 2]])  # Shifted right

    # Failure: Apply a symmetry operation that results in overlap
    failed_walk = np.array([[1, 1], [1, 2], [2, 2], [2, 1]])  # Fold back on itself

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
    plot_walk(initial_walk, ax1, 'Initial Walk')
    plot_walk(successful_walk, ax2, 'Successful Step', success=True)
    plot_walk(failed_walk, ax2, 'Failed Step', success=False)

    plt.show()

if __name__ == '__main__':
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def apply_rotation(path, pivot, angle):
    """Apply a rotation around a pivot point in the path."""
    rad = np.deg2rad(angle)
    rotation_matrix = np.array([
        [np.cos(rad), -np.sin(rad)],
        [np.sin(rad),  np.cos(rad)]
    ])
    rotated_path = path.copy()
    relative_coords = path[pivot:] - path[pivot]
    rotated_coords = (rotation_matrix @ relative_coords.T).T
    rotated_path[pivot:] = rotated_coords + path[pivot]
    return rotated_path

def check_self_avoiding(path):
    """Check if the path is self-avoiding."""
    _, unique_indices = np.unique(path, axis=0, return_index=True)
    return len(unique_indices) == len(path)

def plot_paths(original, modified, title):
    """Plot the original and modified paths."""
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))
    axs[0].plot(original[:, 0], original[:, 1], 'o-', label='Original Path')
    axs[1].plot(modified[:, 0], modified[:, 1], 'o-', label='Modified Path', color='green' if check_self_avoiding(modified) else 'red')
    for ax in axs:
        ax.legend()
        ax.grid(True)
        ax.set_aspect('equal')
    axs[0].set_title('Original Path')
    axs[1].set_title(title)
    plt.show()

# Define the initial path
path = np.array([[0, 0], [1, 0], [1, 1], [2, 1], [2, 2]])

# Choose a pivot and an angle
pivot_index = 2  # Pivot at the third point
angle = 90       # 90 degree rotation

# Apply the rotation
rotated_path = apply_rotation(path, pivot_index, angle)

# Check if the path is still self-avoiding
if check_self_avoiding(rotated_path):
    plot_paths(path, rotated_path, 'Successful Rotation')
else:
    plot_paths(path, path, 'Failed Rotation - Path Overlaps')
```
